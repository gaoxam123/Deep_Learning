```python
from typing import List, Tuple
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def pred_and_plot(model, image_path, class_names, image_size=(224, 224), transform=None, device='cuda'):
    img = Image.open(image_path)

    if transform:
        image_transform = transform

    else:
        image_transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            normalize
        ])

    model.to(device)

    model.eval()
    with torch.no_grad():
        transformed_image = image_transform(img).unsqueeze(0)
        logits = model(transformed_image.to(device))

        preds = torch.softmax(logits, dim=1)
        labels = torch.argmax(preds, dim=1)

        plt.figure()
        plt.imshow(img)
        plt.title(f"Pred: {class_names[labels]} | Prob: {logits}")
        plt.axis(False)
```

```python
def download_data(source, destination, remove_source):
    data_path = Path("data/")
    image_path = data_path / destination
    
    if image_path.is_dir():
        print(f"{image_path}")

    else:
        image_path.mkdir(parents=True, exist_ok=True)

        target_file = Path(source).name
        with open(data_path / target_file, "wb") as f:
            request = requests.get(source)
            f.write(request.content)

        with zipfile.ZipFile(data_path / target_file, "r") as zip_ref:
            zip_ref.extractall(image_path)
        
        if remove_source:
            os.remove(data_path / target_file)

    return image_path
```
